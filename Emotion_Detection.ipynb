##Convert string to integer

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
emotion['Mood']=le.fit_transform(emotion['Mood'])
emotion['Gender']=le.fit_transform(emotion['Gender'])
emotion['Language']=le.fit_transform(emotion['Language'])
emotion['Age Group']=le.fit_transform(emotion['Age Group'])
emotion['Speaker ID']=le.fit_transform(emotion['Speaker ID'])
emotion['MFCCs']=le.fit_transform(emotion['MFCCs'])
```

```python
emotion.head()
```

##Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
y=emotion['Mood']
x=emotion.drop('Mood',axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

##Standerd scaller

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
```

##linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

##r2 score

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

##random forest clasifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

# Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
